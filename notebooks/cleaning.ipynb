```python
import os
import pandas as pd
```

```python
file_path = os.path.join('..', 'data', 'raw', 'Airbnb_Open_Data.csv')
```

```python
df = pd.read_csv(file_path, low_memory=False)
```

```python
df
```

```python
print(f'Dataset Dimensions: {df.shape[0]:,} rows and {df.shape[1]} columns')
```

```python
print(f'Data Technical Breakdown: {df.info()}')
```

```python
df.describe()
```

```python
total_duplicates = df.duplicated(subset=['id']).sum()
```

```python
print(f'Total Duplicates: {total_duplicates:,} rows')
```

```python
rows_before = df.shape[0]
df.drop_duplicates(subset=['id'], inplace=True)
rows_after = df.shape[0]
```

```python
print(f'No. of rows before dropping duplicates: {rows_before:,}')
print(f'No. of rows after dropping duplicates: {rows_after:,}')
```

```python
null_counts = df.isnull().sum()
null_percentage = (null_counts / len(df)) * 100
```

```python
missing_df = pd.DataFrame({'Missing Counts': null_counts, 'Missing Percentage': null_percentage})
```

```python
missing_df[missing_df['Missing Counts'] > 0].sort_values(by='Missing Counts', ascending=False)
```

```python
#license column has 99% missing values, so we can drop it
df.drop(columns=['license'], inplace=True)
```

```python
df['house_rules'] = df['house_rules'].fillna('No rules specified')
```

```python

missing_reviews_context = df[df['reviews per month'].isnull()]['number of reviews'].value_counts()
print(missing_reviews_context)
```

```python
median_review_rate = df['reviews per month'].median()
```

```python
df.loc[(df['reviews per month'].isnull()) & (df['number of reviews'] == 0), 'reviews per month'] = 0.0
```

```python
df.loc[(df['reviews per month'].isnull()) & (df['number of reviews'] > 0), 'reviews per month'] = median_review_rate
```

```python
# Filled reviews per month with 0 for listings with no reviews and median for listings with reviews but missing review rate
missing_reviews_context = df[df['reviews per month'].isnull()]['number of reviews'].value_counts()
print(missing_reviews_context)
```

```python
# Fill missing values in the 'country' column with 'United States'
df['country'] = df['country'].fillna('United States')
```

```python
df['country code'].isnull().sum()
```

```python
# Fill missing values in the 'country code' column with 'US'
df['country code'] = df['country code'].fillna('US')
```

```python
df['neighbourhood group'].unique()
```

```python
df['neighbourhood group'] = df['neighbourhood group'].str.strip().str.title()
```

```python
neighbourhood_group_mapping = {
    'Brookln': 'Brooklyn',
    'Manhatan': 'Manhattan'}
```

```python
#Cleaned the neighbourhood group column by stripping whitespace, converting to title case, and replacing misspelled values with correct ones
df['neighbourhood group'] = df['neighbourhood group'].replace(neighbourhood_group_mapping)
```

```python
df['neighbourhood group'].unique()
```

```python
df.dtypes
```

```python
df['price'] = df['price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
df['service fee'].str.contains('$').sum()
```

```python
df['service fee'] = df['service fee'].str.replace('$', '', regex=False)
```

```python
df['service fee'] = pd.to_numeric(df['service fee'], errors='coerce')
```

```python
df.dtypes
```

```python
df.loc[(df['price'].isnull()) & (df['service fee'].isnull()), ['price', 'service fee']]
```

```python
complete_financials = df[df['price'].notnull() & df['service fee'].notnull() & (df['price'] > 0)]

fee_ratios = complete_financials['service fee'] / complete_financials['price']

print(fee_ratios.value_counts().head())
```

```python
df.loc[(df['price'].isnull()) & (df['service fee'].notnull()), ['price']] = df['service fee']*5
```

```python
df['price'].isnull().sum()
```

```python
df = df.dropna(subset=['price', 'service fee'], how='all')
```

```python
df['price'].isnull().sum()
```

```python
negative_count = (df['minimum nights'] < 0).sum()
```

```python
print(f"Number of rows with negative minimum nights: {negative_count}")
```

```python
df.loc[df['minimum nights'] < 0, 'minimum nights'] = 1
```

```python
df['minimum nights'].min()
```

```python
df['minimum nights'].nlargest(10)
```

```python
(df['minimum nights'] > 365).sum()
```

```python
df.loc[df['minimum nights'] > 365, 'minimum nights'] = 365
```

```python
print("---  FINAL DATASET HEALTH PROFILE ---")
print(f"Total Rows: {df.shape[0]:,}")
print(f"Total Columns: {df.shape[1]}")
print("\n---  DATA TYPES & NULL COUNTS ---")
print(df.info())

print("\n---  FINANCIAL SUMMARY SANITY CHECK ---")
print(df[['price', 'service fee', 'reviews per month']].describe().round(2))
```
